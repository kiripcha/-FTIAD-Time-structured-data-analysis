# –õ–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∞—è –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç—å

–ï—Å–ª–∏ $S(t)$ —è–≤–ª—è–µ—Ç—Å—è —Ü–µ–Ω–æ–π –±–∞–∑–æ–≤–æ–≥–æ –∞–∫—Ç–∏–≤–∞ (–∞–∫—Ü–∏–∏, –∏–Ω–¥–µ–∫—Å–∞), —Ç–æ  –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç—å –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ $\Delta t$:

$$
r_{\Delta t} (t) =  \frac{S(t + \Delta t) - S(t)}{S(t)};$$

–¢–∞–∫ –∫–∞–∫
$$  \dfrac{S(t+1) - S(t)}{S(t)} \approx \ln S(t+1) - \ln S(t), $$
—Ç–æ –Ω–∞ –ø—Ä–∞–∫—Ç–∏–∫–µ —á–∞—Å—Ç–æ —Ä–∞–±–æ—Ç–∞—é—Ç —Å –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–æ–π –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç—å—é:

$$ \ln S(t+1) - \ln S(t). $$
1. –ü–ª—é—Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–∏ –ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å —Ü–µ–Ω–æ–π —Å–∞–º–æ–≥–æ –∞–∫—Ç–∏–≤–∞ - –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è. –°—Ä–∞–≤–Ω–∏–≤–∞—Ç—å –∞–±—Å–æ–ª—é—Ç–Ω—ã–µ —Ü–µ–Ω—ã –¥–≤—É—Ö –∞–∫—Ç–∏–≤–æ–≤ –º–µ–Ω–µ–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ, —á–µ–º —Å—Ä–∞–≤–Ω–∏–≤–∞—Ç—å –∏—Ö –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–∏
2. –†–∞—Å—á–µ—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —Ä–∞–∑–Ω—ã—Ö –∞–∫—Ç–∏–≤–æ–≤ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–µ–µ –ø—Ä–æ–≤–æ–¥–∏—Ç—å –Ω–∞ –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç—è—Ö, –∞ –Ω–µ –Ω–∞ —Ü–µ–Ω–∞—Ö –∞–∫—Ç–∏–≤–∞.



–ú—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∏–µ –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–∏:



1. –õ–æ–≥–Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç—å: –µ—Å–ª–∏ –º—ã –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ–º, —Ç—á–æ —Ü–µ–Ω—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –ª–æ–≥–Ω–æ—Ä–º–∞–ª—å–Ω–æ (—Å—Ç–∞–Ω–¥–∞—Ä–Ω–∞—è –º–æ–¥–µ–ª—å –¥–ª—è —Ä—ã–Ω–∫–∞), —Ç–æ $\log(1 + r_i)$ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ:
$$    1 + r_t = \frac{S(t+1)}{S(t)} = \exp^{\log(\frac{S(t+1)}{S(t)})}. $$
2. –í —Å–ª—É—á–∞–µ –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–µ–π, –±–ª–∏–∑–∫–∏—Ö –∫ –Ω—É–ª—é, –º—ã –º–æ–∂–µ–º –±–µ–∑–±–æ–ª–µ–∑–Ω–µ–Ω–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —á–∏—Å—Ç—ã–µ –ø—Ä–∏—Ä–∞—â–µ–Ω–∏—è, –ø–æ—Ç–æ–º—É —á—Ç–æ:
$$    \log(1 + r) \approx r , r \ll 1 $$
3. –ê–¥–¥–∏—Ç–∏–≤–Ω–æ—Å—Ç—å –ø–æ –≤—Ä–µ–º–µ–Ω–∏:
—Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º n –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω—ã—Ö —Å–¥–µ–ª–æ–∫. –ù–∞–∫–æ–º–ø–ª–µ–Ω–Ω–∞—è –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç—å:
$$
    \displaystyle (1 + r_1)(1 + r_2)  \cdots (1 + r_n) = \prod_i (1+r_i)
$$

–î–∞–Ω–Ω–æ–π —Ñ–æ—Ä–º—É–ª–æ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –Ω–µ—É–¥–æ–±–Ω–æ, —Ç–∞–∫ –∫–∞–∫ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö –≤–µ–ª–∏—á–∏–Ω –Ω–µ —è–≤–ª—è–µ—Ç—Å—è –Ω–æ—Ä–º–∞–ª—å–Ω–æ–π —Å–ª—É—á–∞–π–Ω–æ –≤–µ–ª–∏—á–∏–Ω–æ–π –≤  –æ–±—â–µ–º —Å–ª—É—á–∞–µ. –û–¥–Ω–∞–∫–æ —Å—É–º–º–∞ —Ç–∞–∫–∏—Ö –Ω–æ—Ä–º–∞–ª—å–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã—Ö –≤–µ–ª–∏—á–∏–Ω –±—É–¥–µ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–π (–æ–Ω–∏ –Ω–µ–∫–æ—Ä—Ä–µ–ª–∏—Ä–æ–≤–∞–Ω—ã, –Ω–∞–ø—Ä–∏–º–µ—Ä,–ø–æ —É—Å–ª–æ–≤–∏—é):
$$
    \log(1 + r_t) = \log(\frac{S(t+1)}{S(t)}) = \log(S(t+1)) - log(S(t)) 
$$

–ü–æ—ç—Ç–æ–º—É –Ω–∞–∫–æ–ø–ª–µ–Ω–Ω—ã–µ –ª–æ–≥ –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–∏ –±—É–¥—É—Ç —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ –∏
$$
    \displaystyle \sum_i \log(1+r_i) = \log(1 + r_1) + \log(1 + r_2)  + \cdots + \log(1 + r_n) = \log(S(n)) - \log(S(0))
$$




–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –µ–∂–µ–¥–Ω–µ–≤–Ω–æ–π –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–æ–π –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–∏ –∞–∫—Ç–∏–≤–∞ –ø—Ä–∏–Ω—è—Ç–æ –Ω–∞–∑—ã–≤–∞—Ç—å –¥–Ω–µ–≤–Ω–æ–π –≤–æ–ª–∞—Ç–∏–ª—å–Ω–æ—Å—Ç—å—é –∞–∫—Ç–∏–≤–∞.

# ARCH(p)
–ú–æ–¥–µ–ª—å **ARCH** –ø–æ—Ä—è–¥–∫–∞ p –∏–º–µ–µ—Ç —Å–ª–µ–¥—É—é—â–∏–π –≤–∏–¥:

–ë—É–¥–µ–º –Ω–∞–∑—ã–≤–∞—Ç—å –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Å–ª—É—á–∞–π–Ω—ã—Ö –≤–µ–ª–∏—á–∏–Ω $‚Ñé = (‚Ñé_n)$  ARCH(ùëù)-–º–æ–¥–µ–ª—å—é, –µ—Å–ª–∏
$$‚Ñé_n = ùúé_n ùúÄ_n,$$


–≥–¥–µ $$ùúé^2_n = ùõº_0 + \sum_{k=1}^p ùõº_ùëñ ‚Ñé^2_{n-k},$$  $ùúÄ = (ùúÄ_n) $ ‚Äî –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å iid —Å–ª—É—á–∞–π–Ω—ã—Ö –≤–µ–ª–∏—á–∏–Ω —Å —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º –Ω–æ—Ä–º–∞–ª—å–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º,


$ùõº_0 > 0, ùõº_ùëñ > 0, ‚Ñé_0 = ‚Ñé_0(ùúî)$ ‚Äî —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞, –Ω–µ –∑–∞–≤–∏—Å—è—â–∞—è –æ—Ç $ùúÄ = (ùúÄ_n)_{n\ge 1}.$


–û–±—ã—á–Ω–æ $‚Ñé_0$ –ø–æ–ª–∞–≥–∞—é—Ç –ª–∏–±–æ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–æ–π, –ª–∏–±–æ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω–æ–π, –¥–ª—è –∫–æ—Ç–æ—Ä–æ–π –≤—Ç–æ—Ä–æ–π
–º–æ–º–µ–Ω—Ç –≤—ã–±–∏—Ä–∞–µ—Ç—Å—è –∏–∑ —Å–æ–æ–±—Ä–∞–∂–µ–Ω–∏–π ¬´—Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω–æ—Å—Ç–∏¬ª –∑–Ω–∞—á–µ–Ω–∏–π $E[h_n^2].$



–ò–∑ —Ñ–æ—Ä–º—É–ª—ã –¥–ª—è $ùúé^2_ùëõ$ –≤–∏–¥–Ω–∞ —è–≤–Ω–∞—è –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –æ—Ç $‚Ñé^2_{ùëõ‚àí1}, \dots , ‚Ñé^2_{ùëõ‚àíùëù}.$ 

–ü—Ä–∏ —ç—Ç–æ–º —è—Å–Ω–æ, —á—Ç–æ
–±–æ–ª—å—à–∏–µ (–º–∞–ª—ã–µ) –∑–Ω–∞—á–µ–Ω–∏—è $‚Ñé^2_{ùëõ‚àíùëò}$ –ø—Ä–∏–≤–æ–¥—è—Ç –∫ –±–æ–ª—å—à–∏–º (–º–∞–ª—ã–º) –∑–Ω–∞—á–µ–Ω–∏—è–º $ùúé^2_{ùëõ}.$ 

–í–æ–∑–Ω–∏–∫–Ω–æ–≤–µ–Ω–∏–µ –∂–µ –±–æ–ª—å—à–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è $‚Ñé^2_ùëõ$ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ $‚Ñé^2_{
ùëõ‚àí1},\dots , ‚Ñé^2_{ùëõ‚àíùëù}$ –±—ã–ª–∏ –º–∞–ª—ã–º–∏, –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç
–∏–∑-–∑–∞ –≤–æ–∑–Ω–∏–∫–Ω–æ–≤–µ–Ω–∏—è –±–æ–ª—å—à–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è $ùúÄ_ùëõ.$ 

–≠—Ç–æ –æ–±—ä—è—Å–Ω—è–µ—Ç —Ç–æ, –ø–æ—á–µ–º—É –Ω–µ–ª–∏–Ω–µ–π–Ω—ã–µ –º–æ–¥–µ–ª–∏ –º–æ–≥—É—Ç –ø–æ–º–æ—á—å –≤ –æ–ø–∏—Å–∞–Ω–∏–∏ —Å–æ–±—ã—Ç–∏–π –Ω–∞–ø–æ–¥–æ–±–∏–µ –∫–ª–∞—Å—Ç–µ—Ä–Ω–æ—Å—Ç–∏, —Ç–æ –µ—Å—Ç—å –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞–Ω–∏—è
–∑–Ω–∞—á–µ–Ω–∏–π –≤ –ø–∞—á–∫–∏ ¬´–±–æ–ª—å—à–∏—Ö¬ª –∏ –ø–∞—á–∫–∏ ¬´–º–∞–ª–µ–Ω—å–∫–∏—Ö¬ª –∑–Ω–∞—á–µ–Ω–∏–π.

ARCH —Ä–∞—Å—à–∏—Ñ—Ä–æ–≤—ã–≤–∞–µ—Ç—Å—è, –∫–∞–∫ –∞–≤—Ç–æ—Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–∞—è –º–æ–¥–µ–ª—å —É—Å–ª–æ–≤–Ω–æ–π –Ω–µ–æ–¥–Ω–æ—Ä–æ–¥–Ω–æ—Å—Ç–∏,
–∏–ª–∏ AutoRegressive Conditional Heteroskedastic. 


–ú–æ–¥–µ–ª—å ARCH(q) ‚Äì —ç—Ç–æ –º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ —Ä—è–¥–∞ –∫–∞–∫ AR(q).

–ú—ã –º–æ–∂–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å PACF –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–µ–π, —á—Ç–æ–±—ã –æ—Ü–µ–Ω–∏—Ç—å –ø–æ—Ä—è–¥–æ–∫ ARCH –º–æ–¥–µ–ª–∏.

```python
# –°–∏–º—É–ª–∏–∏—Ä—É–µ–º ARCH(1)

import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline

sp.random.seed(12345)
n=1000 # n is the number of observations
n1=100 # we need to drop the first several observations
n2=n+n1 # sum of two numbers
a=(0.1,0.3) # ARCH (1) coefficients alpha0 and alpha1, see Equation

errors=sp.random.normal(0,1,n2)
t=sp.zeros(n2)
t[0]=sp.random.normal(0,sp.sqrt(a[0]/(1-a[1])),1)
for i in range(1,n2-1):
    t[i]=errors[i]*sp.sqrt(a[0]+a[1]*t[i-1]**2)
plt.title('ARCH (1) process')
y=t[n1-1:-1] # drop the first n1 observations
x=range(n)
plt.plot(x,y)
```

# GARCH (Generalized ARCH)

GARCH(p,q):


$$ h_n = \sigma_n \varepsilon_n,$$

$$\sigma_n^2 = a_0+ \sum_{i=1}^q a_i h_{n-i}^2 + \sum_{j=1}^{p} b_{j} \sigma_{n-j}^2.$$
    
   –û—Å–Ω–æ–≤–Ω–æ–µ –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–æ GARCH(ùëù, ùëû)-–º–æ–¥–µ–ª–µ–π –ø–µ—Ä–µ–¥ ARCH(ùëù)-–º–æ–¥–µ–ª—å—é, —Å–æ—Å—Ç–æ–∏—Ç –≤ –ø–æ–¥–±–æ—Ä–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –º–æ–¥–µ–ª–∏. –ù–∞ –ø—Ä–∞–∫—Ç–∏–∫–µ –ø–µ—Ä–∏–æ–¥–∏—á–µ—Å–∫–∏ –º–æ–∂–µ—Ç –æ–∫–∞–∑–∞—Ç—å—Å—è —Ç–∞–∫, —á—Ç–æ –ø—Ä–∏ –ø–æ–¥–≥–æ–Ω–∫–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö –º–æ–¥–µ–ª—è–º–∏ ARCH(ùëù) –ø–∞—Ä–∞–º–µ—Ç—Ä ùëù —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è —Å–ª–∏—à–∫–æ–º –±–æ–ª—å—à–∏–º (—á—Ç–æ —É—Å–ª–æ–∂–Ω—è–µ—Ç –∞–Ω–∞–ª–∏–∑ –º–æ–¥–µ–ª–∏), –≤ —Ç–æ –≤—Ä–µ–º—è –∫–∞–∫ –ø—Ä–∏ –ø–æ–¥–≥–æ–Ω–∫–µ GARCH(ùëù, ùëû)-–º–æ–¥–µ–ª—è–º–∏ –º–æ–∂–Ω–æ –æ–≥—Ä–∞–Ω–∏—á–∏–≤–∞—Ç—å—Å—è –Ω–µ–±–æ–ª—å—à–∏–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ ùëù –∏ ùëû.

GARCH ‚Äì —ç—Ç–æ ARMA-–ø—Ä–æ—Ü–µ—Å—Å –¥–ª—è –¥–∏—Å–ø–µ—Ä—Å–∏–∏ —Å–µ—Ä–∏–∏.

```python
import scipy as sp
sp.random.seed(12345)
n=1000 # n is the number of observations
n1=100 # we need to drop the first several observations
n2=n+n1 # sum of two numbers
alpha=(0.1,0.3) # GARCH (1,1) coefficients alpha0 and alpha1, see

beta=0.2
errors=sp.random.normal(0,1,n2)
t=sp.zeros(n2)
t[0]=sp.random.normal(0,sp.sqrt(a[0]/(1-a[1])),1)
for i in range(1,n2-1):
    t[i]=errors[i]*sp.sqrt(alpha[0]+alpha[1]*errors[i-1]**2+beta*t[i-1]**2)
y=t[n1-1:-1] # drop the first n1 observations
plt.title('GARCH (1,1) process')
x=range(n)
plt.plot(x,y)
```

$$ y_t = \dfrac{P_t - P_{t-1}}{P_{t-1}} \approx \ln P_t - \ln P_{t-1} $$

–†—è–¥ $y_t$ –æ—Ç—Ä–∞–∂–∞–µ—Ç —Ç–∏–ø–∏—á–Ω—ã–µ ¬´—Å—Ç–∏–ª–∏–∑–æ–≤–∞–Ω–Ω—ã–µ —Ñ–∞–∫—Ç—ã¬ª, –ø—Ä–∏—Å—É—â–∏–µ —Ñ–∏–Ω–∞–Ω—Å–æ–≤—ã–º –¥–∞–Ω–Ω—ã–º:

- Fat tails ‚Äî —ç–∫—Å—Ç—Ä–µ–º–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ—è–≤–ª—è—é—Ç—Å—è —á–∞—â–µ, —á–µ–º –ø—Ä–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–∏ (QQ-plot $y_t$ –∏ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞)
- –ö—Ä–æ–º–µ —Ç–æ–≥–æ, –≤ –¥–∞–Ω–Ω—ã—Ö –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É–µ—Ç leverage effect (–Ω–µ –æ—Ç—Ä–∞–∂–µ–Ω–æ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–∞—Ö) ‚Äî –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ –∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∏–∑–º–µ–Ω–µ–Ω–∏—è –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–∏ –ø–æ —Ä–∞–∑–Ω–æ–º—É –≤–ª–∏—è—é—Ç –Ω–∞ –≤–æ–ª–∞—Ç–∏–ª—å–Ω–æ—Å—Ç—å

    

```python
returns = 100 * price['Adj Close'].pct_change().dropna()
squared_returns = returns ** 2
```

```python
from pandas.plotting import autocorrelation_plot

plt.figure(figsize=(15, 15))

plt.subplot(3, 1, 1)
plt.title('–î–æ—Ö–æ–¥–Ω–æ—Å—Ç—å ‚Äî $y_t$')
returns.plot()

plt.subplot(3, 1, 2)
plt.title('$y_t^2$')
squared_returns.plot()

ax=plt.subplot(3, 1, 3)
plt.title('ACF $y_t^2$')
autocorrelation_plot(squared_returns, ax=ax)
plt.grid()


plt.subplots_adjust(hspace=0.4, wspace=0.1)
plt.show()
```

```python
from scipy.stats import probplot
# import matplotlib.mlab as mlab
from scipy.stats import norm

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ $y_t$ –∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ $y_t$')
returns.hist(bins=200, density=True)
mu = returns.mean()
sigma = returns.std()
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
# plt.plot(x, mlab.normpdf(x, mu, sigma), ls='--', c='red')


y = norm.pdf(x, mu, sigma)
plt.plot(x, y, "r--")
plt.show()



plt.grid()

plt.subplot(1, 2, 2)
probplot(returns, dist="norm", plot=plt)
plt.title('QQ-plot $y_t$')
plt.show()
```

```python
import statsmodels.formula.api as smf
import statsmodels.tsa.api as smt
import statsmodels.api as sm
import scipy.stats as scs
import statsmodels.stats as sms
#from arch import arch_model

import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline

import warnings
warnings.simplefilter("ignore")
```

```python
start = dt.datetime(2014,12,31)
end = dt.datetime(2017,12,31)
data1 = web.get_data_yahoo('^GSPC', start=start, end=end)


data = data1['Adj Close']
# log returns
lrets = np.log(data/data.shift(1)).dropna()
```

```python
def tsplot(y, lags=None, figsize=(15, 10), style='bmh'):
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):    
        fig = plt.figure(figsize=figsize)
      
        layout = (3, 2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))
        qq_ax = plt.subplot2grid(layout, (2, 0))
        pp_ax = plt.subplot2grid(layout, (2, 1))
        
        y.plot(ax=ts_ax)
        ts_ax.set_title('Time Series Analysis Plots')
        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.05)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.05)
        sm.qqplot(y, line='s', ax=qq_ax)
        qq_ax.set_title('QQ Plot')        
        scs.probplot(y, sparams=(y.mean(), y.std()), plot=pp_ax)

        plt.tight_layout()
    return
```

```python

def _get_best_model(TS):
    best_aic = np.inf 
    best_order = None
    best_mdl = None

    pq_rng = range(5) # [0,1,2,3,4]
    d_rng = range(2) # [0,1]
    for i in pq_rng:
        for d in d_rng:
            for j in pq_rng:
                try:
                    tmp_mdl = smt.ARIMA(TS, order=(i,d,j)).fit(
                        method='mle', trend='nc'
                    )
                    tmp_aic = tmp_mdl.aic
                    if tmp_aic < best_aic:
                        best_aic = tmp_aic
                        best_order = (i, d, j)
                        best_mdl = tmp_mdl
                except: continue
    print('aic: {:6.5f} | order: {}'.format(best_aic, best_order))                    
    return best_aic, best_order, best_mdl

```

```python
TS = lrets.dropna()
TS.index = pd.DatetimeIndex(TS.index.values,
                               freq=TS.index.inferred_freq)
res_tup = _get_best_model(TS)
```

```python
order = res_tup[1]
model = res_tup[2]

```

```python
tsplot(model.resid, lags=30)
```

```python
# –§–∏—Ç –º–æ–¥–µ–ª–∏

p_ = order[0]
o_ = order[1]
q_ = order[2]


am = arch_model(model.resid, p=p_, o=o_, q=q_, dist='StudentsT')
res = am.fit(update_freq=5, disp='off')
print(res.summary())
```

```python
#–ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –æ—Å—Ç–∞—Ç–∫–∏

tsplot(res.resid, lags=30)
```
